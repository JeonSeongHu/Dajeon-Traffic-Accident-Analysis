```python
df = pd.read_csv("weather_concat.csv", index_col=0)
df.dropna(inplace=True)
df['Time'] = pd.to_datetime(df['Time'], format='%Y-%m-%d %H:%M:%S')
df['just_time'] = df['Time'].dt.time
df['just_date'] = df['Time'].dt.date
```

```python
print(df.groupby("just_date").count()["Time"].mean(), df.groupby("just_date").count()["Time"].std())
```

```python
df['month_year'] = df['Time'].dt.to_period('M')
tmp_list = list(df.groupby('month_year'))[:-1]
ttest_list = list()
print(len(tmp_list))
ttest_df = []
mean_, std_, num_= [], [], []
for tmp in tmp_list:
    print(tmp[0], end=" ")
    ser = tmp[1].groupby("just_date").count()["Time"]
    ttest_list.append(ser.to_list())
    print(ser.std(), ser.mean(), len(ser))
    std_.append(ser.std()), mean_.append(ser.mean()), num_.append(len(ser))
    
    ser = pd.DataFrame(ser)
    ser["month"] = tmp[0]
    ttest_df.append(ser)
```

```python
ttest_df = pd.concat(ttest_df)
```

```python
ttest_df
ttest_df['month'] = ttest_df['month'].astype(str)
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('Time ~ C(month)', ttest_df).fit()
anova_lm(model)
```

```python
print(scipy.stats.shapiro(ttest_df[ttest_df["month"]=='2020-12']["Time"]))
print(scipy.stats.shapiro(ttest_df[ttest_df["month"]=='2021-01']["Time"]))
print(scipy.stats.shapiro(ttest_df[ttest_df["month"]=='2021-02']["Time"]))
print(scipy.stats.shapiro(ttest_df[ttest_df["month"]=='2021-03']["Time"]))

scipy.stats.bartlett(
    ttest_df[ttest_df["month"]=='2020-12']["Time"],
    ttest_df[ttest_df["month"]=='2021-01']["Time"],
    ttest_df[ttest_df["month"]=='2021-02']["Time"] )
```

```python
from statsmodels.sandbox.stats.multicomp import MultiComparison
import scipy.stats

comp = MultiComparison(ttest_df["Time"],ttest_df["month"])
result = comp.allpairtest(scipy.stats.ttest_ind, method='bonf')
result[0]
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd = pairwise_tukeyhsd(ttest_df["Time"],ttest_df["month"], alpha=0.05)
hsd.summary()
```
